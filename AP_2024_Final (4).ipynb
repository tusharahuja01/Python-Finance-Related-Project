### Question 16
Given a series of trading records for a portfolio, write a script to calculate the maximum drawdown rate and date.

The maximum drawdown is the largest peak-to-trough decline in portfolio value during the trading period.

````python
import pandas as pd

# Create a DataFrame with trading records
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
             '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'],
    'PortfolioValue': [100000, 102000, 101500, 104000, 103000, 99000, 97000, 98000, 101000, 102500]
}
trading_records = pd.DataFrame(data) 
````

````python
# Converting Date column to datetime format
trading_records['Date'] = pd.to_datetime(trading_records['Date'])

# Calculating the cumulative maximum portfolio value
trading_records['CumulativeMax'] = trading_records['PortfolioValue'].cummax()

# Calculating the drawdown as the difference between the cumulative max and the portfolio value
trading_records['Drawdown'] = trading_records['PortfolioValue'] - trading_records['CumulativeMax']

# Finding the maximum drawdown and the corresponding date
max_drawdown_row = trading_records.loc[trading_records['Drawdown'].idxmin()]
max_drawdown = max_drawdown_row['Drawdown']
max_drawdown_date = max_drawdown_row['Date']

# Printing the results
print(f"Maximum Drawdown: {max_drawdown}")
print(f"Date of Maximum Drawdown: {max_drawdown_date.strftime('%Y-%m-%d')}")
````

### Question 19

You are provided with a dataset containing the following columns:
- **model**: Car model (e.g., SUV, Sedan, Truck, Coupe).
- **units_sold**: Number of units sold for each transaction.
- **msrp**: Manufacturer's Suggested Retail Price.
- **sold_price**: The price at which the car was sold.
#### Tasks:
1. **Total Units Sold**: Calculate the total number of units sold for each car model.
2. **Average Sold Price**: Compute the average sold price for each car model.
3. **Average Discount Rate**:
   - Calculate the discount rate using the formula:
 $$
 \text{Discount Rate} = \frac{\text{MSRP} - \text{Sold Price}}{\text{MSRP}} \times 100
 $$
   - Compute the average discount rate for each car model.

The output should look like this:
```yaml
         units  avg_price  avg_discount_rate
model                                       
Coupe      200    27000.0               10.0
Sedan      250    20000.0               20.0
SUV        400    30000.0               14.3
Truck      300    25000.0               10.7
```

````python
import pandas as pd

# Data
data = {
    'model': ['SUV', 'Sedan', 'Truck', 'Coupe', 'SUV', 'Truck', 'Sedan', 'Coupe'],
    'units_sold': [200, 150, 300, 100, 200, 0, 100, 100],
    'msrp': [35000, 25000, 28000, 30000, 35000, 28000, 25000, 30000],
    'sold_price': [30000, 20000, 25000, 27000, 30000, 25000, 20000, 27000]
}
 
df = pd.DataFrame(data)
````

````python
# Calculating discount rate
df['discount_rate'] = ((df['msrp'] - df['sold_price']) / df['msrp']) * 100

# Grouping by car model and calculating required metrics
result = df.groupby('model').agg(
    units=('units_sold', 'sum'),
    avg_price=('sold_price', 'mean'),
    avg_discount_rate=('discount_rate', 'mean')
)

# Rounding the avg_discount_rate to 1 decimal place
result['avg_discount_rate'] = result['avg_discount_rate'].round(1)

# Printing the result
print(result)
````

### Quesiton 22 Cointegration Test (15 points)
Implement a function that tests for cointegration between the two stocks using the coint function from the statsmodels library. Explain how you would use the result to determine if the pair is suitable for trading.

````python
import statsmodels
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint

def test_cointegration(stock_data, ticker1, ticker2, threshold=0.05):
    # Extracting the closing prices of the two stocks
    series1 = stock_data[ticker1]
    series2 = stock_data[ticker2]
    
    # Performing cointegration test using the coint function
    score, p_value, _ = coint(series1, series2)
    
    # Printing the results of the cointegration test
    print(f"Cointegration Test Result for {ticker1} and {ticker2}:")
    print(f"t-statistic: {score}")
    print(f"p-value: {p_value}")
    
    # Determining if the pair is suitable for trading
    if p_value < threshold:
        print("The pair is cointegrated and suitable for trading.")
    else:
        print("The pair is not cointegrated and not suitable for trading.")

# Example usage
df = load_stock_data(['KO', 'PEP'], '2012-01-01', '2024-12-01')  
test_cointegration(df, 'KO', 'PEP')
````

### Quesiton 23 Calculating the Spread (20 points)

To calculate the spread, we use linear regression to estimate the coefficient for the linear combination of the two stocks. This method, known as the Engle-Granger approach, allows us to determine the appropriate weight for constructing the spread. Write a script to calculate the spread and generate a plot of the result.

````python
import yfinance as yf
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Loading stock data for KO and PEP
def load_stock_data(tickers, start_date, end_date):
    data = yf.download(tickers, start=start_date, end=end_date)['Close']
    # Remove rows with NaN values
    data = data.dropna()
    return data

# Calculating spread using Engle-Granger approach (Linear Regression)
def calculate_spread(stock_data, ticker1, ticker2):
    # Extract closing prices of the two stocks
    series1 = stock_data[ticker1]
    series2 = stock_data[ticker2]
    
    # Add constant for intercept in regression model
    X = sm.add_constant(series2)
    
    # Performing linear regression to find the coefficient (beta)
    model = sm.OLS(series1, X).fit()
    
    # Get the beta coefficient (slope) for the regression
    beta = model.params[1]
    print(f"Beta coefficient for {ticker1} and {ticker2}: {beta}")
    
    # Calculate the spread using the formula: Spread = Stock1 - Beta * Stock2
    spread = series1 - beta * series2
    
    return spread

# Function to plot the spread
def plot_spread(spread, ticker1, ticker2):
    plt.figure(figsize=(12, 6))
    plt.plot(spread, label=f"Spread ({ticker1} - Beta * {ticker2})")
    plt.title(f"Spread Between {ticker1} and {ticker2}")
    plt.xlabel("Date")
    plt.ylabel("Spread Value")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
tickers = ['KO', 'PEP']
df = load_stock_data(tickers, '2012-01-01', '2024-12-01')  # Load data for KO and PEP
spread = calculate_spread(df, 'KO', 'PEP')  # Calculate the spread
plot_spread(spread, 'KO', 'PEP')  # Plot the spread
````
